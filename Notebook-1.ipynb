## The purpose of this dataset is to analys of all the movies and TV shows that was available on Netflix from 2018 to 2021.

```sql
SELECT TOP 10 *
FROM netflix_data
ORDER BY date_added DESC;
```

**United States vs other countries.**

```sql
SELECT type, date_added, country
FROM netflix_data
WHERE type = 'movie' AND country = 'united states';

SELECT type, date_added, country
FROM netflix_data
WHERE type = 'movie' AND country NOT LIKE '%united states%'; 
```

**Family and kids categoryÂ "All ages - G"**

```sql
SELECT
type, title, country, duration, rating
FROM netflix_data   
WHERE rating = 'G'
ORDER BY duration DESC; 
```

**Favorite movies.**

```sql
SELECT *
FROM netflix_data   
WHERE title IN ('shutter island', 'pulp fiction', 'the book of eli', 'fallen', 'the silence', 'bird box');
```

**Favorite TV shows.**

```sql
SELECT *
FROM netflix_data   
WHERE title IN ('manifest', 'the sinner',' dead to me', 'the minions of midas', 'the innocent', 'the invisible guest')
```
